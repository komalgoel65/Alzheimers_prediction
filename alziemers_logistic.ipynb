Data Pre Processing

seperating features and target

```python
x = dt.drop(columns=['Diagnosis','DoctorInCharge','Ethnicity','PatientID','EducationLevel','CholesterolLDL','CholesterolHDL','CholesterolTriglycerides','Confusion','ADL'],axis=1).values
y=dt['Diagnosis'].values
```

Splitting the data into training and testing data

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.1,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
scaler=  StandardScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
print(x_train)
```

```python
x_test
```

```python
x
```

```python
from sklearn.linear_model import LogisticRegression
classifier= LogisticRegression(random_state=1)
classifier.fit(x_train, y_train)
```

```python
print(classifier.predict(scaler.transform([[82,1,36.22309879636211,0,4.19289550965694,6.381502407159495,7.971127068871205,9.521026998926784,0,1,0,0,0,0,120,93,271.84144940568734,8.984758990788695,6.946428266027044,0,1,0,0,0,1]])))
```

```python
y_pred= classifier.predict(x_test)
np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm= confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```

```python
input_data = (82,1,36.22309879636211,0,4.19289550965694,6.381502407159495,7.971127068871205,9.521026998926784,0,1,0,0,0,0,120,93,271.84144940568734,8.984758990788695,6.946428266027044,0,1,0,0,0,1)

# changing the input_data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the array as we are predicting for one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

# standardize the input data
std_data = scaler.transform(input_data_reshaped)
print(std_data)

prediction = classifier.predict(std_data)
print(prediction)

if (prediction[0] == 0):
  print('The person is healthy')
else:
  print('The person is having alzheimers')
```
